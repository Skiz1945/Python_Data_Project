```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

---

## ðŸŸ© Calculate Mean Salary (1.26.1)

```python
df['salary_year_avg'].mean()
```

---

## ðŸŸ© Calculate Median Salary (1.26.2)

```python
df['salary_year_avg'].median()
```

---

## ðŸŸ¨ Average Hourly Salary by Job Title (1.26.3)

```python
df.groupby('job_title_short')['salary_hour_avg'].mean()
```

---

## ðŸŸ¨ Job with Lowest Salary (1.26.4)

```python
lowest_hourly_salary = df['salary_hour_avg'].idxmin()
```

```python
df.iloc[lowest_hourly_salary]

#df.iloc[lowest_hourly_salary][['job_title_short', 'salary_hour_avg']]  ------ way to see only 'job_title_short' and 'salary_hour_avg' -------
```
