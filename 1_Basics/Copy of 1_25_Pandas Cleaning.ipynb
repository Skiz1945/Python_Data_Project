```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

---

## ðŸŸ© Sort by Salary (1.25.1)

```python
df.sort_values(by='salary_year_avg', ascending=False, inplace=True)         # "ascending = False" to descending order
```

```python
df.iloc[0:10]
```

---

## ðŸŸ© Create Remote Column (1.25.2)

```python
df['is_remote'] = df.job_work_from_home == True
```

```python
df.info()
```

```python
df
```

---

## ðŸŸ© Extract Year from Date (1.25.3)

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])            # first convert 'job_posted_date' to a pandas datetime with pd.to_datetime

df['job_posted_year'] = df.job_posted_date.dt.year                      # assing new column 'job_posted_year' to converted value from 'job_posted_date' using .dt.year

df.head(5)
```

---

## ðŸŸ¨ Drop Missing Salaries (1.25.4)

```python
df.dropna(subset=['salary_year_avg'], inplace=True)

df.head(5)
```
