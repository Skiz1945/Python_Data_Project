## ðŸŸ© Average Salary (1.16.2)

```python
# Code Needed
salaries = [95000, 120000, 105000, 90000, 130000]

def average_salary():
  return sum(salaries) / len(salaries)

average_salary()
```

## ðŸŸ¨ Salary Statistics (1.16.3)

```python
# Code Needed
salaries = [95000, 120000, 105000, 90000, 130000]

def salary_statistics():
  return {'min': min(salaries), 'max': max(salaries), 'avg': sum(salaries) / len(salaries)}

salary_statistics()
```

## ðŸŸ¥ Job Posting Summary (1.16.4)

```python
# Code Needed
def job_posting_summary():
  total_posting = len(job_postings)
  total_salary = sum(posting['salary'] for posting in job_postings)                # sum salary for every posting in job_postings
  average_salary = total_salary / total_posting
  unique_locations = list(set(posting['location'] for posting in job_postings))
  return {
      'total_postings': total_posting,
      'average_salary': average_salary,
      'unique_locations': unique_locations
  }

job_postings = [
    {'title': 'Data Scientist', 'location': 'New York', 'salary': 95000},
    {'title': 'Data Analyst', 'location': 'San Francisco', 'salary': 85000},
    {'title': 'Machine Learning Engineer', 'location': 'New York', 'salary': 115000}
]

job_posting_summary()
```
