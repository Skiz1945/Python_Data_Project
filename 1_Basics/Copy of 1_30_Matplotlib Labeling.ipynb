```python

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Bar Chart with Labels (1.30.1)

```python
locations_count = df['job_country'].value_counts().head(5)
```

```python
plt.bar(locations_count.index, locations_count.values, color='blue')
plt.xlabel('Country')
plt.ylabel('Number of Postings')
plt.title('Number of Jobs per Country')
plt.xticks(rotation=45, ha='right')
```

---

## ðŸŸ¨ Horizontal Bar Chart with Labels (1.30.2)

```python
salary_per_title = df.groupby('job_title_short')['salary_year_avg'].mean()
salary_per_title = salary_per_title.sort_values(ascending=True)
salary_per_title
```

```python
plt.barh(salary_per_title.index, salary_per_title.values)
plt.xlabel('Average Salary')
plt.title('Average Salary by Job Title')
```

---

## ðŸŸ¥ Line Chart with Labels (1.30.3)

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_month'] = df['job_posted_date'].dt.month
months = df.groupby('job_posted_month')['salary_year_avg'].mean().sort_index()
```

```python
plt.plot(months.index, months.values, color='red', label='Average Salary')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job Month')
plt.legend()
```
