## 游릳 Define Class (1.2.1.4)

```python
job_title = "Data Analyst"
job_location = "United States"
job_salary = 90000

class JobPost:
    def __init__(self, title, location, salary):
        self.title = title                            #atributes
        self.location = location                      #atributes
        self.salary = salary                          #atributes
    def display_info(self):
        return print(f"JOB:       {self.title}\nLOCATION:  {self.location}\nSALARY:    ${self.salary:,.0f}")   #method  (function in class)

JobPost(job_title, job_location, job_salary)
```

```python
job_1 = JobPost(job_title, job_location, job_salary)

print(job_1.title)
print(job_1.location)
print(job_1.salary)
```
