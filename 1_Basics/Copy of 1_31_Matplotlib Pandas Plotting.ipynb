```python

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Job Posting Location Bar Plot (1.31.1)

```python
job_posting_location = df['job_via'].value_counts().head(5)
```

```python
job_posting_location.plot(kind='bar')
```

---

## ðŸŸ© Job Postings Trend Line Plot (1.31.2)

```python
df['monthly_job_counts'] = df['job_posted_date'].dt.to_period('M')
```

```python
job_posting_trend = df['monthly_job_counts'].value_counts().sort_index()
```

```python
job_posting_trend
```

```python
job_posting_trend.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Trend of Job Postings Over Time')
```

---

## ðŸŸ¨ Top Job Locations (1.31.3)

```python
job_locations = df['job_location'].value_counts().head(5)
```

```python
job_locations = job_locations.sort_values(ascending=True)
```

```python
job_locations
```

```python
job_locations.plot(x='job_locations', y='counts', kind='barh')
plt.xlabel('Count of Jobs')
plt.ylabel('Job Location')
plt.title('Number of Job Postings by Job Locations')
```
