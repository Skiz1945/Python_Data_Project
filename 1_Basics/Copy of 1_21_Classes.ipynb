## ðŸŸ© Create JobPosting Class (1.21.1)

```python
# Create the JobPostings class in this code cell

class JobPosting:
  def __init__(self, title, company, location):                  # Use the __init__ method to initialize the attributes of the class.
    self.title = title
    self.company = company
    self.location = location
```

```python
# Code Needed
job = JobPosting('Data Scientist', 'Tech Innovations', 'New York')

print(job.title)
print(job.company)
print(job.location)
```

## ðŸŸ¨ Convert to Dictionary (1.21.2)

```python
# Create the JobPostings class in this code cell

class JobPosting_1:
  def __init__(self, title, company, location, salary):                  # Use the __init__ method to initialize the attributes of the class.
    self.title = title
    self.company = company
    self.location = location
    self.salary = salary

  def to_dict(self):
    return {
        'title': self.title,
        'company': self.company,
        'location': self.location,
        'salary': self.salary
    }
```

```python
# Code Needed
job = JobPosting_1('Data Scientist', 'Tech Innovations', 'New York', 120000)

print(job.to_dict())
```

## ðŸŸ¥ Compare Job Salaries (1.21.3)

```python
# Create the JobPostings class in this code cell

class JobPosting_2:
  def __init__(self, title, company, location, salary):                  # Use the __init__ method to initialize the attributes of the class.
    self.title = title
    self.company = company
    self.location = location
    self.salary = salary

  def compare_salary(self, other_job):
    if self.salary > other_job.salary:
      return self.title
    else:
      return other_job.title


```

```python
# Code Needed
job1 = JobPosting_2('Data Scientist', 'Tech Innovations', 'New York', 120000)
job2 = JobPosting_2('Data Analyst', 'Data Driven Co', 'San Francisco', 100000)

print(job1.compare_salary(job2))
```
