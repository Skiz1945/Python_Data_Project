```python
import ast
import pandas as pd 
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()       
```

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month      # create new column where number of month
```

```python
df_DA_explode = df_DA.explode('job_skills')
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)  # fill_value use to fill NaN values with 0
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()    # use to create new -row- to store sum of each column

df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
df_DA_pivot = df_DA_pivot.drop('Total')
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')
plt.title('Top 5 Skills for Data Analysts per Month')
plt.xlabel('Count')
plt.ylabel('')
plt.show()
```
