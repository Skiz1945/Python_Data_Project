```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Merge US Jobs and Monthly Postings (2.7.1)

```python
df['job_posted_month'] = df['job_posted_date'].dt.to_period('M')
```

```python
df_us_jobs = df[df['job_country'] == 'United States']
```

```python
df_monthly_postings = df.groupby('job_posted_month').size()
df_monthly_postings
```

---

```python
df_monthly_postings = df_monthly_postings.reset_index(name='postings_count')
df_monthly_postings
```

```python
merged_df = pd.merge(df_us_jobs, df_monthly_postings, on='job_posted_month', how='right')
merged_df
```

## 🟨 Merge Non-Remote Jobs and Salary (2.7.2)

```python
df = df.drop(columns='job_posted_month')
```

```python
df_not_remote_jobs = df[df['job_work_from_home'] == False].copy()
df_not_remote_jobs
```

```python
df_not_remote_salary = df_not_remote_jobs.groupby('job_title_short')['salary_year_avg'].mean().reset_index()
df_not_remote_salary
```

```python
df_not_remote_salary.rename(columns={'salary_year_avg': 'salary_not_remote_avg'}, inplace=True)   # important, change column name!
df_not_remote_salary
```

```python
merged_df_1 = pd.merge(df_not_remote_jobs, df_not_remote_salary, on='job_title_short', how='right')
merged_df_1.head()
```

## 🟨 Merge Company Salary and Remote Count (2.7.3)

```python
df_company_salary = df.groupby('company_name')['salary_year_avg'].mean()
df_company_salary = df_company_salary.reset_index()
df_company_salary
```

```python
df_company_remote = df[df['job_work_from_home'] == True]
df_company_remote
```

```python
df_company_remote = df.groupby('company_name').size()
df_company_remote = df_company_remote.reset_index(name='remote_count')
df_company_remote
```

---

```python
merged_df_2 = pd.merge(df_company_salary, df_company_remote, on='company_name', how='inner')
merged_df_2.head()
```
