```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## ðŸŸ© Health Insurance Pie Chart (2.14.1)

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_plot = df_DA['job_health_insurance'].value_counts()
```

```python
df_plot
```

```python
plt.pie(df_plot, labels=['No', 'Yes'], startangle=90, autopct='%1.1f%%')
plt.title('Proportion of Data Analyst Jobs Mentioning Health Insurance')
plt.show()
```

---

## ðŸŸ¨ Job Schedule Distribution (2.14.2)

```python
df_ft_pt = df[(df['job_schedule_type'] == 'Full-time') | (df['job_schedule_type'] == 'Part-time')].copy()
```

```python
job_schedule_count = df_ft_pt['job_schedule_type'].value_counts()
```

```python
plt.pie(job_schedule_count, labels=job_schedule_count.index, startangle=90, autopct='%1.1f%%')
plt.title('Job Schedule Full-time vs Part-time')
```

## ðŸŸ¥ Degree Mentioned Pie Chart (2.14.3)

```python
df_US = df[df['job_country'] == 'United States'].copy()
```

```python
job_no_degree_mention_count = df_US['job_no_degree_mention'].value_counts()
```

```python
job_no_degree_mention_count
```

```python
plt.pie(job_no_degree_mention_count, explode=(0, 0.1), labels=['Degree Mentioned', 'No Degree Mentioned'], startangle=90, autopct='%1.1f%%')
plt.title('Distribution of Jobs Mentioning Degrees for US Job Postings')
```
